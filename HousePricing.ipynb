```python
import pandas as pd

import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt

from math import ceil
```

```python
url = 'https://raw.githubusercontent.com/renatocorreia-rmcm/housing-price/refs/heads/main/housing_price_dataset.csv'
df = pd.read_csv(url)  # load raw csv from github to dataframe object here

df.head()
```

```python
cols = [c for c in df.columns if c not in ['Price', 'Neighborhood_Rural', 'Neighborhood_Suburb', 'Neighborhood_Urban']]  # take of reference and categoric variables
n = len(cols)
ncols = 2  # quantos plots por linha
nrows = ceil(n / ncols)
```

```python
fig, axes = plt.subplots(nrows=nrows, ncols=ncols)

axes = axes.flatten() # transformar axes em lista plana

for ax, col in zip(axes, cols):
    sns.scatterplot(data=df, y=col, x='Price', ax=ax, s=10)

plt.tight_layout()
plt.show()
```
