# Housing Price Model

## SetUp

### Dependencies

```python
import pandas as pd

import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt

from math import ceil
```

### Load DataFrame

```python
url = 'https://raw.githubusercontent.com/renatocorreia-rmcm/housing-price/refs/heads/main/housing_price_dataset.csv'
df = pd.read_csv(url)  # load raw csv from github to dataframe object here

df.head()  # test integrity
```


## Exploratory Data Analysis

```python
# Select only numeric columns
cols = df.select_dtypes(include='number').columns
```

### Histograms

```python
# Iterate through the columns and create a histogram for each
for col in cols:
  plt.hist(df[col], bins=30)
  plt.title(col)
  plt.show()
```

### Boxplots

```python
for c in cols:
  plt.boxplot(df[c])
  plt.title(c)
  plt.show()
```

## Correlation

```python
df.corr().round(2)
```
